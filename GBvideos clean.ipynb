```python
# Load the standard array library
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
# Load the plotting package
import matplotlib.pyplot as plt
# Tell plotting package to show plots in the notebook
%matplotlib inline
# Make plots a little more fancy.
plt.style.use('fivethirtyeight')
```

```python
# Load the standard table manipulation package
import pandas as pd
```

```python
df = pd.read_csv('GBvideos.csv')
df
```

```python
relevant_data = df[['trending_date','title','channel_title','category_id','publish_time','tags','views','likes','dislikes','comment_count']]
relevant_data
```

```python
with open('GB_category_id.json', 'r') as f:
    categories_dict = json.load(f)

items = categories_dict["items"]

data = json_normalize(items)

categories = data[['id','snippet.title']]

cat_id = relevant_data['category_id']
```

```python
for i in np.arange(len(categories)):
       cat_id.replace(int(categories.iloc[i,0]),categories.iloc[i,1],inplace=True)

cat_id.replace(29,'Nonprofits & Activism',inplace=True)
```

```python
#changing the date format to date time to easily sort through the dates and for easier selection of dates
relevant_data['trending_date']=pd.to_datetime(relevant_data['trending_date'],format='%y.%d.%m')
```

We can modify the values in cell 6 to add or remove columns.

```python
relevant_data.rename(columns={'category_id':'category'},inplace=True)
relevant_data
```

```python
relevant_data.to_csv('Clean_GBvideos.csv',index=False)
```
