```python
# Load the standard array library
import numpy as np
# Load the plotting package
import matplotlib.pyplot as plt
# Tell plotting package to show plots in the notebook
%matplotlib inline
# Make plots a little more fancy.
plt.style.use('fivethirtyeight')
# Load the standard table manipulation package
import pandas as pd
```

```python
df = pd.read_csv('Clean_GBvideos.csv')
df
```

```python
#Calculating the number of times each category is in the list
categories = df['category']
total = categories.value_counts()
total
```

```python
total.plot(kind='barh')
plt.title('Number of Times on Trending (11.17 - 06.18)')
plt.ylabel('Categories')
```

```python
#We can infer that the music and entertainment categories make up a large amount of the trending list.
```

```python
#Calculating average category percentage for the above period
avg_percentage = (categories.value_counts()/categories.value_counts().sum())*100
table_avg=pd.DataFrame(data=avg_percentage)
table_avg.columns=['Percentage']
table_avg
```

```python
#Removing categories with less then 1 percentage
revised=avg_percentage.drop(['Autos & Vehicles','Travel & Events','Nonprofits & Activism','Shows'])
```

```python
revised.plot(kind='pie',figsize=(4,4),fontsize=10,startangle=30,radius=1.5)
```

```python
#Finding Out top 3 categories which have the most likes,dislikes,comments,views
#Likes
likes = df.groupby('category')['likes'].sum()
likes=likes.sort_values(ascending=True)
likes=likes.drop(likes.index[:13])
likes.plot(kind='barh')
```

```python
#Dislikes
dislikes = df.groupby('category')['dislikes'].sum()
dislikes=dislikes.sort_values(ascending=True)
dislikes=dislikes.drop(dislikes.index[:13])
dislikes.plot(kind='barh')
```

```python
#Comments
comcount = df.groupby('category')['comment_count'].sum()
comcount=comcount.sort_values(ascending=True)
comcount=comcount.drop(comcount.index[:13])
comcount.plot(kind='barh')
```

```python
#Views
views = df.groupby('category')['views'].sum()
views=views.sort_values(ascending=True)
views=views.drop(views.index[:13])
views.plot(kind='barh')
```

```python
#Comparing properties of different categories such as like/dislike ratio,comments etc
#Like/Dislike Ratio:
ldratio = df.groupby('category')['likes'].agg('sum') / df.groupby('category')['dislikes'].agg('sum')
#Dropping shows here as it has very low amount of videos and thus has a very high ratio
ldratio = ldratio.drop(['Shows'])
ldratio = ldratio.sort_values(ascending=True)
ldratio.columns = ['category','ratio']
ldratio.plot(kind='barh')
```

```python
#Ratio of Comments to number of Views - Finding out which category of videos attracts the most discussion/feedback
comratio = df.groupby('category')['comment_count'].sum()/df.groupby('category')['views'].sum()
comratio = comratio.sort_values(ascending=True)
comratio.columns = ['category','ratio']
comratio.plot(kind='barh')
```
