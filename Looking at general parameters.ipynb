```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
import pandas as pd
```

```python
#All the videos on trending in GB from November 2017 to June 2018
df = pd.read_csv('GBvideos.csv')
df.head()
```

```python
#First I had a look at how many times a channel was on trending. This could include consecutive days.
channel = df['channel_title']
channel
```

```python
type(channel)
```

```python
#It looks like the most common channels to find on trending were tv shows and talk show channels.
times_trending = df['channel_title'].value_counts()
times_trending
```

```python
#I looked at the most/least viewed videos
col_list = ['title', 'views']
```

```python
df_views = df[col_list]
```

```python
df_views.head()
```

```python
sort_views = df_views.sort_values('views',ascending=False)
sort_views
```

```python
#Looking at the mean, standard deviation, maximum and minimum for the views, likes, dislikes an comment count. For each value the standard deviation is massive, showing very little correlation. This meant that it was pointless to compare these variables when making an algorithm.
df55 = df.describe()
df55
```

```python
#I looked at the most/least liked videos
col_list2=['title', 'likes']
df_likes = df[col_list2]
df_likes.head()
```

```python
#Very surprised some videos got 0 likes and still went on trending. This just confirmed my previous conclusion.
sort_likes = df_likes.sort_values('likes',ascending=False)
sort_likes
```

```python
#I looked at the most/least disliked videos
col_list3=['title', 'dislikes']
df_dislikes = df[col_list3]
df_dislikes.head()
```

```python
#Again, some videos on trending got 0 dislikes, this is very surprising.
sort_dislikes = df_dislikes.sort_values('dislikes',ascending=False)
sort_dislikes
```

```python
#I looked like/dislike ratio.
div = df['likes']/df['dislikes']
div.head()
```

```python
type (div)
```

```python
df = df.copy()
df['like_dislike_ratio'] = div.values
```

```python
col_list4=['title', 'like_dislike_ratio']
df_ratio = df[col_list4]
df_ratio.head()
```

```python
#As some videos have 0 likes or 0 dislikes I got N/A and inf values so I had to drop these.
sort_ratio = df_ratio.sort_values('like_dislike_ratio',ascending=False)
sort_ratio
```

```python
df2 = sort_ratio.dropna()
```

```python
df3 = df2.replace([np.inf, -np.inf], np.nan)
```

```python
df4 = df3.dropna()
```

```python
#Finally, the like/dislike ratio seems to have no impact on the video going on trending either.
df4
```

```python
#To conclude, it seems like the number of views, likes, dislikes, like-dislike ratio and comment count has no impact on whether the video gets on trending or not so I can't use any of this in the algorithm. We did find out that the minimum number a views a video has got on trending is 815. 
```
