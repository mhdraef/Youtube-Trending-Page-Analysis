```python
# Load the standard array library
import numpy as np
# Load the plotting package
import matplotlib.pyplot as plt
# Tell plotting package to show plots in the notebook
%matplotlib inline
# Make plots a little more fancy.
plt.style.use('fivethirtyeight')
# Load the standard table manipulation package
import pandas as pd
```

```python
df3 = pd.read_csv('mostviewedvids6.csv')
df3
```

```python
sort2 = df3.sort_values('Views',ascending=False)
sort2
```

```python
df = pd.read_csv('GBvideos.csv')
df.head()
```

```python
col_list = ['title', 'channel_title', 'views', 'trending_date']
```

```python
df_views = df[col_list]
df_views
```

```python
sort_views = df_views.sort_values('views',ascending=False)
sort_views
```

```python
df_nodup = sort_views.drop_duplicates(subset='title', keep='first', inplace=False)
df_nodup
```

```python
#merging the two dataframes. If one video was in one dataframe and not in the other then it wouldn't show up below. All 6 videos showed up.
mergeddf = pd.merge(df_nodup, sort2, on=['title'], how='inner')
mergeddf
```

```python
sort_views100 = mergeddf.sort_values('views',ascending=False)
sort_views100
```

```python
#Yes, all 6 of the most popular videos between November 2017 and June 2018 were on trending, which is what you would expect. It would have been nicer to find an anomaly so we could find out why that wasn't on trending.
```
